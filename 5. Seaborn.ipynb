```python
import seaborn as sns
import numpy as np
# Advanced library for Data visualization
```

```python
# https://seaborn.pydata.org  Documentation
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Bar plot
sns.barplot(x = 'sex', y = 'total_bill', data=tips, estimator = np.std)
# Here, we can specify the estimator but in counter plot we can't.
```

```python
# Counter plot
# Just similar to box plot
sns.countplot(x='sex', data = tips)
```

```python
# Box plot
sns.boxplot(x = "day", y= "total_bill" , data=tips, hue = 'smoker')
```

```python
# Violin plot
# sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = False)
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
# Dis plot
# sns.distplot(tips['total_bill'])
sns.histplot(tips['total_bill'], kde=True, bins=40)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# Joint Plot
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind='scatter')
```

```python
# Pair plot
sns.pairplot(tips, hue='sex')
```

```python
# Corelation [-1, 1]
tips_corr = tips.corr()
tips_corr
```

```python
# Heat map
sns.heatmap(tips_corr, annot=True)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flight_pivot = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(flight_pivot, cmap = "coolwarm")
```
