```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
```

```python
themes = plt.style.available
print(themes)
plt.style.use('seaborn')
```

```python
plt.plot(x, y1, color='red', label="Apple", marker='^')
plt.plot(x, y2, color='green', label="Kiwi", linestyle="dashed", marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()

plt.show()
```

```python
prices = np.array([1, 2, 3, 4])**2
print(prices)
```

```python
plt.plot(prices)


plt.show()
```

```python
# Scatter plots
plt.scatter(x, y1)
plt.scatter(x, y2)

# Adjust the Size of any Plot
plt.figure(figsize = (6,6))

plt.show()
```

```python
# Bar graphs
x_coordinates = np.array([0,1,2])*2
# plt.bar([0,1,2],[10, 20, 15])
plt.style.use("dark_background")
plt.bar(x_coordinates,[10, 20, 15], width=0.5, label="Current Year", tick_label=["Gold", "Platinum", "Silver"])
plt.bar(x_coordinates + 0.5,[20, 10, 12], width=0.5, color="red", label="Next Year")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0, 40)
plt.legend()

plt.show()
```

```python
# Pie Chart

subject = ["Maths", "Chemistry", "Physics", "English"]
weightage = [20, 10, 15, 5]

# plt.pie(weightage, labels = subject
plt.pie(weightage, labels = subject, explode=(0,0,0.1,0), autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
# Histogram
# Std Normal Distribution -> mean = 0 and std = 1

xsn = np.random.randn(100)
sigma = 5
u = 60
X = np.round(xsn*sigma + u)
X2 = np.round(xsn*5 + 40) 
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X, alpha = 0.8, label="Maths")
plt.hist(X2, alpha = 0.8, label="Science")
plt.ylabel("Prob/FreqCount of Students")
plt.xlabel("Marks Range")
plt.title("Histogram")
plt.show()
```
